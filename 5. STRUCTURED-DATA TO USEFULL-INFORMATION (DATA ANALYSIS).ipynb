#### **MatPlotLib.PyPlot**
---

Uma biblioteca de gr√°ficos e visualiza√ß√£o de dados, muito utilizada na √°rea cient√≠fica e matem√°tica.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Global Population over the Years
population = pd.read_csv("https://gist.githubusercontent.com/sjl/2394233/raw/a1d835f4d65357ff6632ad924e7180c5916b37e2/world-population.csv")
population = pd.DataFrame(population["population"].values, index=population["year"])
population = population.loc[(population.index > 0)] # Gets only after Christ years
population.index.name = "Years"
population.columns = ["Population"]

# NBA Players Statistics in 2019
players = pd.read_csv("https://raw.githubusercontent.com/dataprofessor/data/master/nba-player-stats-2019.csv")
```

O pyplot guarda o gr√°fico como um endere√ßo de mem√≥ria espec√≠fica. Ou seja, no momento que √© criado, o endere√ßo fica ocupado e n√£o permite intera√ß√µes m√∫ltiplas. Partindo desse endere√ßo, √© poss√≠vel fazer as altera√ß√µes e chamadas de fun√ß√µes, incluindo a de mostrar.

Os recursos podem ser usados inclusive juntos com outras bibliotecas de plottagem.

```python
plt.figure(figsize=(8, 4)) # Size (width, height), figure always before plot
plt.plot(population) # Create Memory Graph
plt.xlim(0, 2020) # (min, max) from "x" visualization
plt.ylim(0, 3000000000) # (min, max) "y" visualization
plt.xlabel("Years") # Label of X
plt.ylabel("Population") # Label of Y
plt.title("Population/Years") # Graph Title
plt.show() # Show the Graph
```

#### **Multiple Graphs**

Os gr√°ficos m√∫ltiplos do Pyplot funcionam de forma semelhante com gr√°fico √∫nico. √â carregado um gr√°fico pro vez e plotado no final.

```python
# Multiple Graphs
plt.subplot(2, 2, 1) # (total rows, total cols, position) of the Graph
plt.subplots_adjust(wspace=0.3, hspace=0.7) # Graphs Spacing in %
plt.plot(population.loc[0:500])
plt.title("First 500 Years Population")
plt.subplot(2, 2, 2)  # (total rows, total cols, position) of the Graph
plt.plot(population.loc[0:1000])
plt.title("First Milenium Population")
plt.subplot(2, 2, 3)  # (total rows, total cols, position) of the Graph
plt.plot(population.loc[1000:2000])
plt.title("First to Second Milenium Population")
plt.subplot(2, 2, 4)  # (total rows, total cols, position) of the Graph
plt.plot(population)
plt.title("Population")
plt.show()
```

#### **Seaborn**

Uma biblioteca baseada no matplotlib s√≥ que com mais detalhamento nos gr√°ficos melhorando sua est√©tica e identifica√ß√£o de informa√ß√µes, por√©m esses recursos consomem mais processamento do que comparado com o matplotlib.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Global Population over the Years
population = pd.read_csv("https://gist.githubusercontent.com/sjl/2394233/raw/a1d835f4d65357ff6632ad924e7180c5916b37e2/world-population.csv")
population = pd.DataFrame(population["population"].values, index=population["year"])
population = population.loc[(population.index > 0)] # Gets only after Christ years
population.index.name = "Years"
population.columns = ["Population"]
mil = []
for i in population.index:
  if i < 1000:
    mil.append("First")
  else:
    mil.append("Second")
population["Milenium"] = mil

# NBA Players Statistics in 2019
players = pd.read_csv("https://raw.githubusercontent.com/dataprofessor/data/master/nba-player-stats-2019.csv")
```

```python
sns.lineplot(x="Years", y="Population", data=population, linewidth=1, marker=".", markersize=10).set(title="Population Over Years", xlabel="Years", ylabel="Population in Billions")
sns.set_theme(style="white", font_scale=0.7) # Font Write
```
