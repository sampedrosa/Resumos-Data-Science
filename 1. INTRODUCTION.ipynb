## **üîº AIRFLOW**
---


o Apache Airflow √© uma ferramenta gratuita e aberta de gerenciamento de fluxo de opera√ß√µes para pipelines de dados. Teve origem em 2014 pela Airbnb para solucionar seus problemas de fluxos de opera√ß√µes, se tornando posteriormente um projeto da Apache. √â utilizado para orquestrar pipelines criadas chamadas de **DAGs** (Grafos Ac√≠clicos Direcionados) com cada grafo sendo uma **Task** que √© compilada por um **Operator**.

image.png

```python
from airflow import DAG
from airflow.operators.python import PythonOperator # Operator to compile the Task
from airflow.operators.python import BranchPythonOperator # Operator to compile and branch into other Tasks
from airflow.operators.bash import BashOperator # Terminal Instruction Operator
from airflow.operators.email import EmailOperator # Operator sends an email
from airflow.providers.http.operators.http import HttpOperator # Operator calls an HTTP endpoint
from airflow.providers.mysql.operators.mysql import MySqlOperator # Operator executes an SQL code

def Task1():
  var = 0
  # Do Something
  return var

def Task2(task_instance):
  var = task_instance.xcom_pull(task_ids='Task1') # Pull Variable from other Task
  if var > 0:
    return 'Task3'
  if var < 0:
    return 'Task4'
  else:
    return 'Task5'

with DAG('Name ID', start_date='Date to Start', schedule_interval='Start Interval', catchup=False) as dag:
  Task1 = PythonOperator(task_id='Task1', python_callable=Task1)
  Task2 = BranchPythonOperator(task_id='Task2', python_callable=Task2)
  Task3 = BashOperator(task_id='Task3', bash_command="Shell Instruction")
  Task4 = BashOperator(task_id='Task4', bash_command="Shell Instruction")
  Task5 = BashOperator(task_id='Task5', bash_command="Shell Instruction")

  Task1 >> Task2 >> [Task3, Task4, Task5]
```
