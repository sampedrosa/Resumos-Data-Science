**Probabil√≠stica (Probabilistic)**
*    Campo da probabilidade. Baseada em c√°lculos de eventos incertos, desconhecidos ou empiricamente impratic√°veis. Podem ser:
 *    **Probabilidade de Frequ√™ncia ou Aleatoriedade**: Quando h√° uma maneira de calcular o resultado e sua previsibilidade matem√°tica. Como jogar uma moeda ideal.

Para a distribui√ß√£o funcionar corretamente √© necess√°rio que cada experimento do valor tenha as mesmas caracter√≠sticas, ou seja, a mesma probabilidade binomial.

Ou seja, um evento com **N** experimentos independentes de probabilidade **P** (True) e probabilidade **Q=1-P** de (False). Assim, temos a seguinte f√≥rmula para calcularmos a  distribui√ß√£o de **P(x)**:

Essa fun√ß√£o √© a chamada Distribui√ß√£o Binomial com formato semelhante da figura abaixo, variando apenas com **N** e **P**. H√° tamb√©m uma Tabela Binomial com os valores j√° definidos.

**Exemplo**: Em uma prova de 20 quest√µes de 5 alternativas (A, B, C, D, E). Procura-se a probabilidade de "chutar" aleatoriamente todas as quest√µes e conseguir 80% de acertos (m√≠nimo 16 acertos). Ou seja:
```
P(X>=16) = P(16) + P(17) + P(18) + P(19) + P(20)
P = 1/5 (20% de Acerto)
Q = 1-P (80% de Erro)
N = 20 (Quest√µes)
-------------------------------
P(16) = 1.30 E-8
P(17) = 7.65 E-10
P(18) = 3.19 E-11
P(19) = 8.39 E-13
P(20) = 1.05 E-14
P(X>=16) = 2,1054 E-9
Pr√≥ximo de uma chance em meio bilh√£o.
```

````python
# Binomial Simulator
# Test with N questions with random probability P of getting the right answer
import numpy as np
X = 0.8 # Minimum Percentage of Right in the Test
P = 0.2 # Correct Answer Probability
N = 20 # Number of Questions per Test

number_of_tests = 0

while(True):
  test = [] # N Questions
  for i in range(N):
    if np.random.random() > P: # (1-P) chances
      test.append(False) # Wrong Answer
    else:                      # P chances
      test.append(True)  # Correct Answer
  if np.mean(test) >= X: # When the Test has X or more Correct Answers
     print(f"One case in {number_of_tests}.")
     break
  number_of_tests = number_of_tests + 1
````

#### **Poisson Distribution**
---

A Distribui√ß√£o de Poisson √© uma distribui√ß√£o discreta para uma probabilidade de um determinado evento ocorrer em um intervalo de tempo ou espa√ßo. A m√©dia dos eventos precisa ser conhecida e serem independentes do tempo passado. Como por exemplo a probabilidade de um acidente de carro acontecer em uma cidade durante um dia.

A f√≥rmula da distribui√ß√£o de probabilidade em fun√ß√£o de **X** com a **M√©dia = Œª** √©:

image.png

**Exemplo**: Em um Maternidade P√∫blica de uma cidade, √© necess√°rio saber quantas vagas de gestantes por dia ser√° o suficiente para suprir a demanda com no m√°ximo 0.5% de chance de faltar vagas durante um dia. Sabe-se que a M√©dia (lambda Œª) √© de 3 gestantes por dia. Como n√£o h√° um limite de gestantes, utilizamos a l√≥gica contr√°ria e tentamos achar um n√∫mero de vagas **V** que supra 99.5% dos dias. Ou seja, a somat√≥ria de zero vagas at√© todas terem sido ocupadas no dia.
```
M√©dia de Gestantes ou Vagas Ocupadas (lambda) = 3

Probabilidade de Faltar Vagas:
P(x) = 0.005 = P(V) + P(V+1) + P(V+2) ...

Probabilidade de N√£o Faltar Vagas:
1-P(x) = 0.995 = P(0) + P(1) + P(2) ... P(V)
```
Para achar **V**, podemos utiizar um algoritmo:

````python
import numpy as np
e = 2.718281
lambd = 3
Q = 0.995

P = 0.0
x = 0
while(P < Q): # P(0) + P(1) + ... P(P = Q)
  Px = (np.power(lambd, x) * np.power(e, -lambd))/ (np.math.factorial(x))
  P = P + Px
  print(f'P(X={x}) = {Px:,.5f} +')
  x = x + 1

print(f'---------------- \nP(X<{x}) = {P:,.5f} and V = {x-1}')
````

Ou seja,
*  P(8) = (3^8)/(e^3 * 8!) = 0.81%
*  P(9) = (3^9)/(e^3 * 9!) = 0.27%
*  P(X<8) = 98.81%
*  P(X>7) = 01.19% > 0.5%
*  P(X<9) = 99.62%
*  P(X>8) = **00.38% < 0.5%**

Ent√£o com 8 vagas na maternidade a chance de n√£o suprir a demanda durante um dia √© menor que a probabilidade m√°xima permitida.

Isso ocorre devido ao comportamento das probabilidades de poisson formando uma distribui√ß√£o. Segue a plotagem do gr√°fico do exemplo formando a Distribui√ß√£o de Poisson:

````python
# Poisson Graph
import seaborn as sns
probs = []

for x in range(20):
  probs.append((np.power(lambd, x) * np.power(e, -lambd))/ (np.math.factorial(x)))

sns.scatterplot(probs)
````

### **‚ñ∂  Confidence**
---

Supomos uma popula√ß√£o de **P** pessoas que podem ser "Favor√°vel" (True) ou "Contra" (False) uma lei. Consideramos que 64% seja "Favor√°vel".

````python
import pandas as pd
import numpy as np
import seaborn as sbr
P = 1000000

population = []

for i in range(P):
  if np.random.random() > 0.64: # 64% will be True
    population.append(False)
  else:
    population.append(True)

votes = np.average(population)
print(f'{votes} from the Population voted True')
````

Queremos realizar uma pesquisa dessa popula√ß√£o, podemos pegar uma amostra (sample) de tamanho **N** em vez dos **P** da popula√ß√£o. Tamb√©m podemos criar **M** testes repetidos, cada um com **N** pessoas.

Utilizamos a Taxa de Erro (ERRORTAX) e a fun√ß√£o errorFunction para tornar a amostragem inclinada um pouco mais √° votar "Contra" (False), j√° que n√£o h√° teste de amostragem perfeito na vida real.

Para uma realiza√ß√£o de pesquisa como essa, √© necess√°rio determinar o n√≠vel de confian√ßa:

**N√≠vel de Confian√ßa**: Uma porcentagem determinada, geralmente maior que 80%, para considerar apenas os testes de amostragem que se aproximam. Ou seja, desconsidera os outliers e cria-se um intervalo de confian√ßa. O **Alpha** √© determinado pela metade de 1 - NC (TL), assim exclui-se a porcentagem Alpha dos outliers maiores e dos menores.

````python
N, M = 20, 100
ERRORTAX = 0.05
TL = 0.95 # Trust Level
alpha = (1 - TL)/2
````

Ap√≥s a defini√ß√£o de **P**, **N**, **M** e o **N√≠vel de Confian√ßa (TL)**, podemos preencher as amostras (samples) e realizar os testes, tamb√©m criamos uma lista de m√©dias para preencher a m√©dia de cada amostra.

Plotamos um Box Chart para visualiza√ß√£o.

````python
sample = [] # size N
means = [] # shape (M, N)

for i in range(M):
  sample = np.random.choice(a=[True, False], size=N, replace=True, p=[0.64, 0.36])
  means.append(sample.mean())

print(f'Average in Samples: {np.average(means)}')
sbr.boxplot(means)
````

---
#### **Simple Confidence (Samples)**
---

Como estamos considerando testes de amostragens, podemos apenas retirar os outliers e obter facilmente com as amostragens restantes o intervalo de confian√ßa e a margem de erro.

**Intervalo de Confian√ßa das Amostragens**: Ap√≥s a determina√ß√£o do n√≠vel de confian√ßa, exclui-se os outliers e o intervalo √© determinado entre o resultado do teste com menor m√©dia e o com maior m√©dia. Ou seja, considerando um n√≠vel de confian√ßa de 95%, excluimos os 5% dos testes que mais fugiram da m√©dia e determinamos o intervalo com os 95% restantes.

**Margem de Erro das Amostragens**: √â a diferen√ßa que delimita o intervalo de confian√ßa. Ou seja, IC (CI) = M√©dia dos Resultados (+/-) Margem de Erro.

````python
simple_means = []
for mean in means:
  simple_means.append(mean) # Using another list to simple confidence

for i in range(int(alpha*M)): # Getting the outliers
  simple_means.pop(simple_means.index(max(simple_means)))
  simple_means.pop(simple_means.index(min(simple_means)))

simpleCI = [min(simple_means)*100, max(simple_means)*100] # Samples Confidence Interval
simpleME = ((max(simple_means) - min(simple_means))/2)*100 # Samples Margin of Error

print(f'Real Population Votes: {votes} \nFOR SIMPLE CONFIDENCE: \nAverage in Samples with Trust Level: {np.average(simple_means)} \nConfidence Interval for Samples: {simpleCI[0]}% to {simpleCI[1]}% \nMargin of Error for Samples: {simpleME}%')
sbr.boxplot(simple_means)
````

Como conslus√£o, temos que a Confian√ßa Simples das Amostragens depende consideravelmente dos valores de **N** e **M** e do **N√≠vel de Confian√ßa**. Quanto maior o n√≠vel de confian√ßa, maior vai ser o intervalo. Quanto maior for N, mais as amostragens v√£o ser pr√≥ximas do real. Quanto maior for M, mais outliers v√£o existir. Um N e M altos melhoram o experimento, por√©m h√° a necessidade de mais pessoas e pode invibializar a inte√ß√£o do uso da amostragem.

#### **Average Confidence**
---

Para m√©dias, √© poss√≠vel usar uma abordagem diferente.

Utilizando o Teorema Central do Limite, podemos plotar um histograma das m√©dias das amostragens que se aproxima de uma distribui√ß√£o normal. Como uma forma de aproximarmos do resultado real, excluiremos os outliers, ou seja, uma porcentagem das extremidades da distribui√ß√£o normal, as mais distantes do pico/m√©dia.

````python
sbr.histplot(means, kde=True)
````



**Erro Padr√£o**: Obtem-se da tabela Z o valor correspondente ao erro padr√£o. O valor √© EP(SE) = Z(1 - Alpha) ou ent√£o EP(SE) = Z*(TL/2) se for uma tabela vers√£o reduzida. Ou seja, se o N√≠vel de Confian√ßa √© 96%, Alpha √© 2% e o valor a ser procurando na tabela Z √© 0.98 ou 0.48 na reduzida Z* que √© igual a 2.055.

**Intervalo de Confian√ßa para M√©dias**: √â definida pela forma seguinte, com o c√°lculo da margem de erro diferente da abordagem simples.

**Margem de Erro para M√©dias**: √â definida na formula seguinte, com o correspondente Z (erro padr√£o) vezes o desvio padr√£o m√©dio dividido pela raiz do tamanho da amostra:

image.png

````python
import scipy.stats as st

std = np.std(means) # Standard Deviation

SE = st.norm.ppf(1-alpha) # getting Z score to be Standard Error
avgME = (SE*std)/np.sqrt(M) # Average Margin of Error

print(f'Real Population Votes: {votes*100}% \nFOR AVERAGE CONFIDENCE: \nConfidence Interval = {np.average(means)*100}% (+/-) {avgME*100}% \nOR {(np.average(means)-avgME)*100}% to {(np.average(means)+avgME)*100}%')
````

√â percept√≠vel que a Confian√ßa pelas M√©dias tende a ter resultados mais interessantes, entretanto depende de um **M** significante para que a distribui√ß√£o possa se assemelhar a uma normal Z.

---
#### **Proporcional Confidence**
---

Para a Confian√ßa de Propor√ß√£o, basta apenas modificar a formula do desvio padr√£o para **œÉ = ‚àö(xÃÑ(1-xÃÑ))**.

````python
import scipy.stats as st

pstd = np.sqrt(np.average(means)*(1-np.average(means))) # Proporcional Standard Deviation

SE = st.norm.ppf(1-alpha) # getting Z score to be Standard Error
avgME = (SE*pstd)/np.sqrt(M) # Average Margin of Error

print(f'Real Population Votes: {votes*100}% \nFOR AVERAGE CONFIDENCE: \nConfidence Interval = {np.average(means)*100}% (+/-) {avgME*100}% \nOR {(np.average(means)-avgME)*100}% to {(np.average(means)+avgME)*100}%')
````

### **‚ñ∂  Statistical Tests**

A Estat√≠stica de Teste ou Testes de Hip√≥teses √© um procedimento de definir uma Hip√≥tese (Hip√≥tese Nula) e uma Contra-Hip√≥tese (Hip√≥tese Alternativa) sobre um conhecimento estat√≠stico. Como por exemplo uma pesquisa de amostragem apontar algo, ent√£o o Teste de Hip√≥tese pode ser usado com outra amostragem sendo a Hip√≥tese o que j√° foi considerado pela pesquisa e a Contra-Hip√≥tese uma suposi√ß√£o se a pesquisa est√° errada por mais ou menos. Utiliza-se uma vari√°vel chamada N√≠vel de Signific√¢ncia.

**N√≠vel de Signific√¢ncia**: Vari√°vel definida com valores baixos. Dependendo do tamanho da amostragem de do valor escolhido, o procedimento pode errar.

**Hip√≥tese**: Hip√≥tese a ser contestada. Exemplo: X = 23%

**Contra-Hip√≥tese**: Hip√≥tese a ser testada. Podendo ser um resultado maior ou menor que a Hip√≥tese. Exemplo: X > 23% ou X < 23%

**Valor de Teste (P)**: Valor obtido pela obten√ß√£o na Tabela Z. Se a Contra-Hip√≥tese sugerir maior que a hip√≥tese, ent√£o P = 1-Z. Caso seja menor, P = Z.

O resultado √© obtido comparando P com o N√≠vel de Signific√¢ncia. Ou seja,
```
NS = N√≠vel de Signific√¢ncia
H0 = Hip√≥tese Nula
H1 = Contra-Hip√≥tese
Se H1 > H0, P = 1-Z, Se NS > P, H1 √© V√°lido e H0 Rejeitado
Se H1 < H0, P = Z,   Se NS < P, H1 √© V√°lido e H0 Rejeitado
```
Dessa maneira, √© poss√≠vel testar estatisticamente as hip√≥teses, desde que a amostragem seja de tamanho consider√°vel e n√£o seja corrompida. H√° duas formas de Teste de Hip√≥tese, uma pela M√©dia e outra pela Propor√ß√£o.

---
#### **Average Hypothesis Test (Z)**
---

Para obter o valor de P em M√©dia, utiliza-se a f√≥rmula e depois calcula-se o Z final pelo resultado na Tabela Z:

image.png

Dessa maneira, √© poss√≠vel calcular **P=1-Z (Maior) ou P=Z (Menor)**.
*  Se (H1 > H0) e (NS > P), ent√£o H1 √© V√°lido e H0 Rejeitado. Caso n√£o, H0 n√£o √© Rejeitado.
*  Se (H1 < H0) e (NS < P), ent√£o H1 √© V√°lido e H0 Rejeitado. Caso n√£o, H0 n√£o √© Rejeitado.

**Exemplo**: Se uma pesquisa aponta que a m√©dia de peso de uma esp√©cie √© 64kg. Entretanto, acredita-se que esteja errado. Logo podemos determinar as hip√≥teses e definir um valor para o N√≠vel de Seguran√ßa (NS).
```
NS = 0.05
H0 |Peso M√©dio = 64kg
H1 |Peso M√©dio > 64kg
H2 |Peso M√©dio < 64kg
```
Obtemos a coleta de informa√ß√µes de uma Amostragem de tamanho N e obtemos:
```
N = 100
M√©dia = 65kg
Desvio Padr√£o = 3
```
Assim, calculamos o Z = Z((Med-H0)/(DP/‚àöN)) = Z((65-64)/(3/‚àö100)) = Z(3,33) = 0.9996
*  Para **H1**: P = 1-Z = 0.0004 < 0.05 Logo, **H1 √© V√°lido e H0 Rejeitado**.
*  Para **H2**: P = Z(Z) = 0.9996 > 0.05 Logo, **H2 n√£o √© V√°lido**.

Conclui-se que a m√©dia de peso √© maior que 64kg.

#### **Proporcional Hypothesis Test (Z)**
---

Para obter o valor de P em Propor√ß√£o, utiliza-se a f√≥rmula e depois calcula-se o Z final pelo resultado na Tabela Z:

image.png

Dessa maneira, √© poss√≠vel calcular **P=1-Z (Maior) ou P=Z (Menor)**.
*  Se (H1 > H0) e (NS > P), ent√£o H1 √© V√°lido e H0 Rejeitado. Caso n√£o, H0 n√£o √© Rejeitado.
*  Se (H1 < H0) e (NS < P), ent√£o H1 √© V√°lido e H0 Rejeitado. Caso n√£o, H0 n√£o √© Rejeitado.

**Exemplo**: Se uma pesquisa aponta que 64% √© favor√°vel. Entretanto, acredita-se que esteja errado. Logo podemos determinar as hip√≥teses e definir um valor para o N√≠vel de Seguran√ßa (NS).
```
NS = 0.05
H0 |Favor√°vel 64%
H1 |Favor√°vel > 64%
H2 |Favor√°vel < 64%
```
Obtemos a coleta de informa√ß√µes de uma Amostragem de tamanho N e obtemos:
```
N = 100
M√©dia = 63%
```
Assim, calculamos o Z = Z((Med-H0)/(‚àö(H0*(1-H0)/N)) = Z((0.63-0.64)/(0.64*0.36/‚àö100)) = Z(-0.0659) = 0.484
*  Para **H1**: P = 1-Z = 0.516 < 0.05 Logo, **H1 n√£o √© V√°lido**.
*  Para **H2**: P = Z(Z) = 0.484 > 0.05 Logo, **H2 √© V√°lido e H0 √© Rejeitado**.

Conclui-se que menos que 63% √© Favor√°vel.

#### **Chi-Square œá2**
---

O Qui-Quadrado √© uma distribui√ß√£o que utiliza testes de hip√≥tese que considera um intervalo aceito para que a vari√°vel em quest√£o n√£o seja desconsiderada de uma fun√ß√£o **F(x)**. Ou seja, essa fun√ß√£o possui uma "margem toler√°vel" de erro baseada no N√≠vel de Signific√¢ncia (alpha) tanto no limite superior e inferior.
```
F(x) = Fun√ß√£o Qui-Quadrado
N√≠vel de Signific√¢ncia = alpha
H0 | N√£o existe diferen√ßa significativa al√©m do acaso
H1 | x > k
H2 | x < k

Ao mostrar que H1 e H2 n√£o invalidam H0, cria-se um intervalo [xi, xf]:
Z(xi) = -alpha e Z(xf) = +alpha
k = x +/- Acaso ou F(x) = F(k) +/- Acaso
```

A f√≥rmula para frequ√™ncia **Observada (Oi)** e **Esperada (Ei)** √©:

image.png

**Exemplo**: Dois grupos de pessoas (A, B) foi testado para uma doen√ßa. O grupo A teve o resultado de 83 positivos e 317 negativos, j√° o grupo B, foi 21 positivos e 79 negativos. Realizando o algoritmo do X^2, temos:

````python
import numpy as np
from scipy.stats import chi2_contingency

alpha = 0.05
A = [83, 317]
B = [19, 81]
test = np.array([A, B])

p = chi2_contingency(test).pvalue

if p > alpha:
  print("No significant difference between A and B")
else:
  print("There is significant difference between A and B")
````

#### **Anova (Analysis of Variance)**
---

A An√°lise de Vari√¢ncia √© uma maneira de aplica√ß√£o de testes em dados estat√≠sticos com hip√≥tese para verificar se h√° signific√¢ncia entre os dados. Ou seja, compara m√©dias e utiliza uma **Hip√≥tese (H0) = N√£o h√° diferen√ßa significativa entre as m√©dias**.

Como os dados podem aparecer em formatos ou em grupos diferentes, h√° v√°rias formas de testes:
*  **Teste de Student:** Apenas entre duas m√©dias (H0 | alpha < Z(P))
*  **Teste F**: Tr√™s ou mais m√©dias (H0 | alpha > Z(P))
*  **Teste de Tukey**: Identifica onde est√° as varia√ß√µes significativas.


**Exemplo**: Um rem√©dio para dor de cabe√ßa √© testado em diversas pessoas separadas por grupos para testar seu funcionamento em horas at√© que a pessoa n√£o tenha mais dor de cabe√ßa. No caso de apenas dois grupos de pessoas, pode-se utilizar o Teste de Student, j√° no caso de mais grupos, usa-se o Teste F. No caso as pessoas s√£o entre homens e mulheres sendo jovens, adultos e idosos, ou seja, 6 Grupos.

Ap√≥s a coleta de dados, cria-se a hip√≥tese H0 para verificar se h√° difere√ßa significativa entre as pessoas pelos devidos grupos. Idealmente o rem√©dio √© feito para ser efici√™nte em todos os tipos de pessoas. Calcula-se Z(P) obtendo a m√©dia das horas coletadas e o desvio padr√£o. Depois √© s√≥ comparar com o N√≠vel de Signific√¢ncia (no caso como √© Teste F (alpha > Z(P))) e descobrir que h√° vari√¢ncia signific√¢ncia entre os grupos ou n√£o.

````python
import pandas as pd
import numpy as np
N = 100 # People Size

treatment = []

for i in range(N):
  treat = []
  if np.random.random() > 0.5:
    treat.append("Male")
  else:
    treat.append("Female")

  if np.random.random() < 0.4:
    treat.append("Young")
  elif np.random.random() < 0.7:
    treat.append("Adult")
  else:
    treat.append("Elder")

  treat.append(int(np.random.random()*10))

  treatment.append(treat)

treatment = pd.DataFrame(treatment)
treatment.columns = ["Gender", "Age", "Hours"]
````

````python
import statsmodels.api as sm
from statsmodels.formula.api import ols

model = ols("Hours ~ Age * Gender", data=treatment).fit()
display(sm.stats.anova_lm(model)) # Higher PR(>F) means that H0 is Valid
````

#### **Linear Regression**
---

Regress√£o Linear √© quando h√° uma equa√ß√£o que relaciona duas vari√°veis de forma linear, ou seja, crescimento ou decrescimento proporcional. Gr√°ficamente podemos tra√ßar uma linha da fun√ß√£o **y=f(x)** que consegue estimar valores de dados n√£o coletados.

Padronizadamente, o **Eixo Y √© o Dependente/Resposta** e o **Eixo X o Independente/Explanat√≥ria**, ou seja, queremos ver a rea√ß√£o de y em fun√ß√£o da varia√ß√£o de x. Uma rela√ß√£o de duas vari√°veis independentes √© chamado de **Colinearidade** e geralmente n√£o gera resultados n√£o confi√°veis ou pouco informativos.

A rela√ß√£o das vari√°veis, mesmo estatisticamente correta, nem sempre indica uma Rela√ß√£o de Efeito e Causa. Como h√° uma quantidade muito grande de possibilidades de correlacionamentos de vari√°veis distintas, a chance de alguma dessas correla√ß√µes serem nada mais que coincid√™ncia √© extremamente prov√°vel.

Mesmo em uma rela√ß√£o de correlacionamento compreens√≠vel, √© necess√°rio analisar se uma vari√°vel est√° de fato afetando outra, uma vez que √© poss√≠vel que seja o contr√°rio. Como por exemplo a rela√ß√£o de Pessoas Altas x Jogar Basquete Bem. As pessoas altas jogam basquete bem ou as pessoas que jogam basquete bem s√£o altas?


**Exemplo**: H√° uma rela√ß√£o direta com os minutos jogados por jogo de um jogador da NBA com a sua quantidade de pontos marcados por jogo. Uma vez que quanto mais ele joga, mais ele marca. Assim, podemos fazer uma Regress√£o Linear desses dados e perceber que gr√°ficamente √© poss√≠vel tra√ßar uma linha.

````python
import pandas as pd
import numpy as np
import seaborn as sns

players_stats = pd.read_csv("https://raw.githubusercontent.com/dataprofessor/data/master/nba-player-stats-2019.csv")
players_stats = players_stats.drop_duplicates("Player")

x = players_stats["MP"] # Minutes Played per Game
y = players_stats["PTS"] # Points per Game

sns.regplot(x=x, y=y, line_kws={"color":"red"})
````

**Covari√¢ncia**: √â a m√©dia da diferen√ßa dos dados com a m√©dia para mais de uma vari√°vel, no caso de uma vari√°vel √© chamada de Vari√¢ncia. Para y=f(x):

image.png

**Correla√ß√£o (œÅ)**: Consideramos a Correla√ß√£o das duas vari√°veis como um n√∫mero **œÅ** que varia entre (-1...1). H√° diferentes formas de se calcular, mas a mais comum √© a de Pearson que √© a covari√¢ncia dividido pela produto dos desvios padr√µes das vari√°veis:

image.png
```
œÅ = 0  | Inexistente (Sem Rela√ß√£o)
œÅ = +1 | Perfeita (Crescimento)
œÅ = -1 | Perfeita Inversa (Decrescimento)
```
**Coeficiente de Determina√ß√£o (œÅ^2)**: Como uma forma mais operacional para trabalhar com porcentagens e n√∫meros absolutos, elevamos a Correla√ß√£o ao quadrado e obtemos o Coeficiente de Determina√ß√£o que varia entre (0...1) ou em quantos porcento de dados pode ser explicado pela rela√ß√£o.
```
œÅ^2 = 1.0       | Rela√ß√£o Perfeita
1.0 > œÅ^2 > 0.7 | Rela√ß√£o √ìtima
0.7 > œÅ^2 > 0.5 | Rela√ß√£o Boa
0.5 > œÅ^2 > 0.3 | Rela√ß√£o Ruim
0.3 > œÅ^2 > 0.0 | Rela√ß√£o Desconsider√°vel
```



Sendo ent√£o a reta definida como uma fun√ß√£o de primeiro grau, temos a Intersec√ß√£o (C) e a Inclina√ß√£o (B):
```
y = f(x) = B*x + C
```
**Inclina√ß√£o (B)**: √â a covari√¢ncia das vari√°veis dividida pela vari√°veis a ser prevista. Ou ent√£o:

image.png

**Intercepta√ß√£o (C)**: Sabendo a f√≥rmula da reta y=f(x) e a Inclina√ß√£o (B ou M), basta achar a constante da Intercepta√ß√£o utilizando o x e y m√©dios na f√≥rmula:

image.png



````python
œÅ = np.corrcoef(x.values, y.values)[0][1] # Gets the XX, XY, YX and the YY Correlations, Gets only the XY or YX
œÅ2 = œÅ*œÅ # Coefficient of Determination

B = œÅ*(np.std(y)/np.std(x)) # Inclination
C = y.mean() - B*x.mean() # Interceptation

print(f'œÅ = {œÅ} \nœÅ^2 = {œÅ2} \ny = f(x) = ({B}*x) + ({C})')
````

**Previs√£o**: Com a f√≥rmula da regress√£o linear, podemos fazer previs√£o para qualquer x de y=f(x). Para dados inferiores ao menor(x) e superiores ao maior(x), n√£o √© recomendado fazer previs√µes por ser uma **Extrapola√ß√£o**.
```
Previs√£o(x) = B*(x) + C
```
**Res√≠duo**: √â a diferen√ßa do dado real com a sua Previs√£o. Os Res√≠duos tendem a ter um comportamento de distribui√ß√£o normal e podem ser analisados dessa forma.
```
y = Previs√£o(x) (+/-) Res√≠duo
```
Segue o Histograma dos Res√≠duos do exemplo tendendo pra uma Distribui√ß√£o Normal:

````python
residues = []

for i in range(len(x.values)):
  residue = (y.values[i] - (B*(x.values[i]) + C))
  residues.append(residue)

sns.histplot(residues, kde=True)
````

#### **Logistic Regression**
---

A Regress√£o Log√≠stica √© semelhante a uma Regress√£o Linear, por√©m a vari√°vel independente/explanat√≥ria tem car√°ter bin√°rio/booleano.
```
y = f(x) = True ou False (Heaviside)
Maior/Menor (x) = Maior/Menor Probabilidade (y)
```
Em vez de uma reta apenas, a fun√ß√£o da Regress√£o Log√≠stica √© mistura de uma fun√ß√£o de primeiro grau com uma fun√ß√£o sigmoide:
```
y(p) = ln(p/(1-p))
y(x) = b1*x + b0
```
Logo, podemos considerar a Regress√£o Log√≠stica **y(p) = y(x)**. O b0 e b1 s√£o estimados apartir de M√°xima Versossimilhan√ßa com m√©todos matem√°ticos variados de otimiza√ß√£o, como Newton-Raphson, Limited-Memory BFGS e Stochastic Average Gradient. Igualando as fun√ß√µes:

image.png

Como divis√£o por zero e logaritmo natural de zero ou negativo s√£o indefini√ß√µes matem√°ticas, p assume um valor aberto entre zero at√© um, enquanto y possui um comportamento sigmoidal e x continua linear. Ou seja, **p indica a probabilidade de um x ser y=True ou y=False**.

image.png



**Exemplo**: N√£o s√£o muitos jogadores da NBA que conseguem uma m√©dia de 15 pontos ou mais por jogo. Considerando a perspectiva booleana de conseguir (True) e n√£o conseguir (False) a marca dos 15 pontos/jogo.

````python
import pandas as pd
import numpy as np
import seaborn as sns

players_stats = pd.read_csv("https://raw.githubusercontent.com/dataprofessor/data/master/nba-player-stats-2019.csv")
players_stats = players_stats.drop_duplicates("Player")
players_stats.loc[players_stats["PTS"] < 15.0, "PTS"] = 0
players_stats.loc[players_stats["PTS"] >= 15.0, "PTS"] = 1

x = players_stats[["MP"]].values # Minutes Played per Game
y = players_stats["PTS"].values # True for PTS equal or more then 15, False otherwise

sns.regplot(x=x, y=y, logistic=True, line_kws={"color":"red"})
````

Dessa maneira, come√ßamos a perceber que a probabilidade pra um jogador ter a m√©dia de mais de 15 pontos por jogo come√ßa a ficar mais prov√°vel do que improv√°vel quando sua m√©dia de minutos por jogo come√ßa a passar dos 30 minutos.

Com a fun√ß√£o definida, √© poss√≠vel fazer Previs√µes com x sendo os minutos e y a probabilidade:

````python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(solver='liblinear', C=10.0, random_state=0).fit(x, y)
# solver can be Newton-Raphson (newton-cg), Limited-Memory BFGS (lbfgs), Stochastic Average Gradient (sag) or Stochastic Average Gradient Accelerated (SAGA)
b0 = lr.intercept_ # Interceptation
b1 = lr.coef_ # # Inclination

MP = 30
prob = lr.predict_proba([[MP]])[:, 1][0]
print(f'Player with {MP} Minutes Played per Game have {prob*100:,.2f}% of Scoring More then 15 PTS')
````

````python
# Uma forma de analisar o gr√°fico √© que existe a distribui√ß√£o logistica com o scatter dos zeros e uns
x_lr = np.linspace(x.min(), x.max(), len(x))
y_lr = 1/(1 + np.power(np.e, -(x_lr*b1 + b0)))

sns.scatterplot(x=x.ravel(), y=y)
sns.lineplot(x=x_lr, y=y_lr.ravel(), color="red")
````

### **‚ñ∂  Time Series**
---

As S√©ries Temporais √© uma rela√ß√£o de uma vari√°vel dependente/resposta com o tempo. Ou seja, um dado coletado vinculado ao seu tempo de coleta, necess√°rio ordem desses valores e por isso sendo chamados de S√©ries. Dessa maneira, a vari√°vel n√£o vai se relacionar com outra vari√°vel, mas com a vari√°vel anterior. Como por exemplo coleta de dados de temperatura em uma cidade, se em uma hora √© coletado 28 graus, a proxima coleta ser√° um valor pr√≥ximo desse (dependente).

A S√©rie pode ser **Est√°cion√°ria** variando em torno de uma mesma m√©dia como a temperatura clim√°tica de um dia ou n√£o est√°cion√°ria como uma bolsa de valores. Al√©m disso, quando uma fun√ß√£o explica exatamente o comportamento da S√©rie, √© chamada de **Determin√≠stica**, j√° quando h√° uma fun√ß√£o estimativa e fatores desconhecidos afetando a vari√°vel, √© chamado de **Estoc√°stico**.

As principais caracter√≠sticas da **Decomposi√ß√£o** de uma S√©rie Temporal s√£o:

*  **Tend√™ncia**: √â o comportamento a longo prazo da S√©rie. Os dados variam entre s√≠, mas se aproxiasse-mos de uma regress√£o linear, ocorreria uma taxa de crescimento ou decrescimento.
*  **Sazonalidade**: Quando h√° ciclos de tempos ou eventos que se repetem. Por exemplo pre√ßos de passagem de avi√£o ao longo de um ano, j√° que aumenta em per√≠odos de f√©rias.
*  **Aleatoriedade**: Na decomposi√ß√£o de uma S√©rie Temporal Estoc√°stico, √© poss√≠vel separar a Tend√™ncia (Trend), a Sazonalidade (Seasonal) e ainda um efeito Aleat√≥rio (Random), tamb√©m chamado de erro, que √© desconhecido, mas foi coletado e pode tamb√©m ser analisado.

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from datetime import datetime

dateparse = lambda dates: datetime.strptime(dates, '%Y-%m-%d')
oilprice = pd.read_csv("https://raw.githubusercontent.com/datasets/commodity-prices/master/data/commodity-prices.csv", usecols=["Date","Crude Oil petroleum"], parse_dates=["Date"], index_col="Date", date_parser=dateparse)
decomp = seasonal_decompose(oilprice) # Decompose the Data Frame in a Time Series

sns.lineplot(oilprice)
````

````python
# Trend
plt.subplot(3, 1, 1)
sns.lineplot(decomp.trend)

oilprice_year = oilprice.groupby([lambda x: x.year]).sum()
plt.subplot(3, 1, 2)
sns.lineplot(oilprice_year)

oilprice_year.index = oilprice_year.index/10
oilprice_year.index = oilprice_year.index.astype(int)
oilprice_decade = oilprice_year.groupby("Date").sum()
oilprice_decade.index = oilprice_decade.index*10

plt.subplot(3, 1, 3)
sns.lineplot(oilprice_decade)
````

````python
# Seasonal
sns.lineplot(decomp.seasonal)
````

````python
# Random/Residual
sns.lineplot(decomp.resid)
````

#### **ARIMA**
---

O ARIMA (AutoRegressive Integrated Moving Average) √© um modelo de previs√£o para s√©ries temporais que substitui os dados coletados com a diferen√ßa entre os seus valores e os dados coletados anteriormente, justamente para ajustar em um modelo mais previs√≠vel e aumentar a depend√™ncia de um dado com a coleta anterior. Al√©m disso o erro de regress√£o √© uma combina√ß√£o linear. Ou seja, temos **ARIMA(p,d,q)**:
*  **AutoRegressivo (AR)**: Rela√ß√£o dos Per√≠odos. **p** √© a Ordem da Regress√£o.
*  **Integra√ß√£o (I)**: Diferencia√ß√£o. **d** √© o Grau de Diferencia√ß√£o.
*  **M√©dia Andante (MA)**: Extra√ß√£o de Erros dos Per√≠odos. **q** √© a Ordem da M√©dia

Os par√¢metros (p,d,q) s√£o encontrados com base em testes dentro da S√©rie Temporal que minimizem as m√©tricas de avalia√ß√£o.

````python
!pip install pmdarima
````

````python
from pmdarima.arima import auto_arima

stepwise = auto_arima(oilprice, m=12, seasonal=True, trace=True, stepwise=True)
print(stepwise.aic()) # Metric Testing the Model (the lower, the best)
````

````python
stepwise.fit(oilprice)
predict = stepwise.predict(n_periods=60)

sns.lineplot(predict)
````

√â percept√≠vel que para um exemplo como o pre√ßo do petr√≥leo ao longo dos √∫ltimos anos, a previs√£o do ARIMA se concentra numa m√©dia com o passar do tempo, uma vez que o exemplo possui muito efeito residual. Entretanto, para casos mais pr√≥ximos da √∫ltima previs√£o de teste, o ARIMA consegue ter alguma relev√¢ncia na previsibilidade.
